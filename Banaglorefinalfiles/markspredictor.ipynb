```python
## program to generate dummy dataset
import pandas as pd

# Sample data
data = {
    'Student ID': range(1, 6),  # Generate student IDs
    'Student Name': ['Student 1', 'Student 2', 'Student 3', 'Student 4', 'Student 5'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Age': [15, 16, 17, 15, 16],
    'Family Size': [4, 3, 5, 4, 4],
    'Parental Education': ['High School', "Bachelor's Degree", "Master's Degree", 'High School', 'High School'],
    'Study Time': ['3 hours', '5 hours', '4 hours', '2 hours', '3 hours'],
    'Absences': [2, 1, 3, 0, 4],
    'Exam 1 Score': [75, 85, 70, 90, 65],
    'Exam 2 Score': [80, 90, 75, 85, 70],
    'Final Grade': [78, 88, 72, 88, 68]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('student_performance.csv', index=False)
```

```python
### code with just two marks test

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Function to load and preprocess the user-input dataset
def load_dataset(file_path):
    df = pd.read_csv(file_path)
    # Perform any preprocessing steps if necessary
    return df

# Function to predict marks using the trained model
def predict_marks(model, df):
    X = df[['Age', 'Family Size', 'Absences', 'Exam 1 Score', 'Exam 2 Score']]
    df['Predicted Marks of Next Test'] = model.predict(X).round().astype(int)
    return df

# Function to display the dataset with actual and predicted marks
def display_dataset(df):
    print(df)

# Main function
def main():
    # Step 1: Allow the user to input the dataset file path
    dataset_path = input("Enter the path to the dataset file: ")

    # Step 2: Load the dataset provided by the user
    dataset = load_dataset(dataset_path)

    # Step 3: Split Data into Training and Testing Sets
    X = dataset[['Age', 'Family Size', 'Absences', 'Exam 1 Score', 'Exam 2 Score']]
    y = dataset['Final Grade']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Step 4: Load the trained model
    model = GradientBoostingRegressor(random_state=42)
    model.fit(X_train, y_train)

    # Step 5: Predict marks based on the input dataset
    dataset_with_predictions = predict_marks(model, dataset)

    # Step 6: Display the dataset with actual and predicted marks
    display_dataset(dataset_with_predictions)

    # Step 7: Evaluate the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Display model accuracy
    print("\nModel Accuracy:")
    print("Mean Squared Error:", mse)
    print("R-squared Score:", r2)

# Execute the main function
if __name__ == "__main__":
    main()
```

```python

### if we input marks of many test instead of one .

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Function to load and preprocess the user-input dataset
def load_dataset(file_path):
    df = pd.read_csv(file_path)
    # Perform any preprocessing steps if necessary
    return df

# Function to predict marks using the trained model
def predict_marks(model, df, exam_score_columns):
    X = df[exam_score_columns]
    df['Predicted Marks of Next Test'] = model.predict(X).round().astype(int)
    return df

# Function to display the dataset with actual and predicted marks
def display_dataset(df):
    print(df)

# Main function
def main():
    # Step 1: Allow the user to input the dataset file path
    dataset_path = input("Enter the path to the dataset file: ")

    # Step 2: Load the dataset provided by the user
    dataset = load_dataset(dataset_path)

    # Step 3: Split Data into Training and Testing Sets
    exam_score_columns = [col for col in dataset.columns if 'Exam' in col]  # Extract exam score columns dynamically
    X = dataset[exam_score_columns]
    y = dataset['Final Grade']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Step 4: Load the trained model
    model = GradientBoostingRegressor(random_state=42)
    model.fit(X_train, y_train)

    # Step 5: Predict marks based on the input dataset
    dataset_with_predictions = predict_marks(model, dataset, exam_score_columns)

    # Step 6: Display the dataset with actual and predicted marks
    display_dataset(dataset_with_predictions)

    # Step 7: Evaluate the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Display model accuracy
    print("\nModel Accuracy:")
    print("Mean Squared Error:", mse)
    print("R-squared Score:", r2)

# Execute the main function
if __name__ == "__main__":
    main()
```
